```python
from google.colab import files
```

```python
import easyocr
import pandas as pd
import os
```

```python
test = pd.DataFrame({'File':[],
                              'Text':[]})
```

```python
reader = easyocr.Reader(['en','pl'])
```

```python
directory = "/content/test_set/"
```

```python
i=0
for filename in os.listdir(directory):
    text = reader.readtext(f'{directory}/{filename}', detail = 0)

    test = test.append({'File': filename,
                          'Text': text}, ignore_index=True)
    print(i)
    i+=1
```

```python
test.to_csv('test.csv') 
files.download('test.csv')
```
